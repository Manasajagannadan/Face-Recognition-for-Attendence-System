```python
import glob
import numpy as np
import os
from PIL import Image
image_list = []
resized_image = []
```

```python
for filename in glob.glob(r'C:\Users\SHANKARA\Downloads\p\*.bmp'):
    print(filename)
    img = Image.open(filename)
    image_list.append(img)
```

```python
for image in image_list:
    #image.show()
    image = image.resize((640, 480))
    resized_image.append(image)
```

```python
for (i, new) in enumerate(resized_image):
    new.save('{}{}{}'.format(r'C:\Users\SHANKARA\Desktop\new', i+1, '.bmp'))
```
